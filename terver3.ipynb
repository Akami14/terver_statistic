# 1.	
Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.



```python
import numpy as np
X = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150],  dtype=np.float64)

print(X)
X_mean = X.sum()/X.shape[0]
print(f'среднее {X_mean}')

sum_temp=0
for el in X:
    sum_temp += (el-X_mean)**2
print(f'дисперсия смещеная - {round(sum_temp/X.shape[0],4)}')    
print(f'дисперсия не смещеная - {round(sum_temp/(X.shape[0]-1),4)}')  
print(f'среднеквадратичное отклонение  - {round(sum_temp/(X.shape[0])**0.5,4)}')  
```

# 2.	
В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?


есть 3 ситуации 
* 2 белых из 1го и 1 из второго соответсвено другие шары не должны быть белыми
* 1 белый из первого и 2 белых из второго соответсвено другие шары не должны быть белыми
* 0 белых из 1го и 3 белых из второго  соответсвено другие шары не должны быть белыми


```python
from math import factorial
def combination(n,m):
    return int(factorial(n))/(int(factorial(m))*int(factorial((n-m))))
```

```python
p1 = (combination(5,2)*combination(3,0)*combination(5,1)*combination(7,3))/(combination(8,2)*combination(12,4))
p2 = (combination(5,1)*combination(3,1)*combination(5,2)*combination(7,2))/(combination(8,2)*combination(12,4))
p3 = (combination(5,0)*combination(3,2)*combination(5,3)*combination(7,1))/(combination(8,2)*combination(12,4))
p=p1+p2+p3
print('Вероятность того, что 3 мяча белые - ', round(p,4))
```

Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?


* Используем формулу вероятности байяса 
* P(H|A)=P(A|H)*P(H)/P(A)
* P(H|A) - вероятность того что сдал счтудент с факультета A или B или С H1 H2 H3
* P(A|H) - вероятность сдачи студента с факультета А B или С H1 H2 H3
* P(A) - полная группа вероятности стача студента с любого факультета 
* P(A)  = P(A|H1)*P(H1) + P(A|H2)*P(H2)+ P(A|H3)*P(H3)
* eсли принять что количество учеников на А или B факультете за x то на С = x+x=2x
следовательно вероятность выбор 1 ученика из А или p(H1)=p(H2) = x/(x+x+2x)=1/4 для С p(H3)=2x/(x+x+2x)=1/2
* P(A) = 0.8*0.25+0.7*0.25+0.9*0.5 = 0.825


* 1) Студент с А факультета
P(H1|A)=P(A|H1)*P(H1)/P(A)
P(H1|A)= 0.8*0.25/0.825 = 0.2424
* 2) Студент с B факультета
P(H1|A)=P(A|H2)*P(H2)/P(A)
P(H1|A)= 0.7*0.25/0.825 = 0.2121
* 3) Студент с C факультета
P(H1|A)=P(A|H3)*P(H3)/P(A)
P(H1|A)= 0.9*0.5/0.825 = 0.5455
